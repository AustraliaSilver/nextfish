```python
# 4. Ch·∫°y th·ª≠ nghi·ªám t√≠ch h·ª£p Hybrid
import subprocess
import os

def test_nextfish():
    engine_path = "./stockfish"
    # model.onnx n·∫±m ·ªü th∆∞ m·ª•c cha c·ªßa src
    model_path = os.path.abspath("../model.onnx")
    
    process = subprocess.Popen(engine_path, stdin=subprocess.PIPE, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
    
    # Giao th·ª©c UCI
    cmds = [
        "uci\n",
        f"setoption name Lc0Policy_ModelPath value {model_path}\
        "setoption name Lc0Policy_Active value true
",
        "isready
",
        "position startpos
",
        "go depth 18
",
        "quit
"
    ]
    
    print("--- ƒêang kh·ªüi ƒë·ªông Nextfish v2.0 Hybrid ---")
    for c in cmds:
        process.stdin.write(c)
        process.stdin.flush()
        
    while True:
        line = process.stdout.readline()
        if not line: break
        print(line.strip())
        if "bestmove" in line: break

test_nextfish()
```
